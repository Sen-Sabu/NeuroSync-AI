```python
from flask import Flask, render_template_string, request
from pyngrok import ngrok
import threading

app = Flask(__name__)

HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>NeuroSync AI</title>
    <style>
        body {
            background: #0f172a;
            color: white;
            font-family: Arial;
            display: flex;
            justify-content: center;
            padding: 40px;
        }

        .container {
            width: 500px;
            background: #1e293b;
            padding: 30px;
            border-radius: 12px;
        }

        h1 {
            color: #38bdf8;
        }

        form {
            display: flex;
            flex-direction: column;
        }

        input, select, button {
            margin-top: 10px;
            margin-bottom: 20px;
            padding: 10px;
            border-radius: 8px;
            border: none;
        }

        button {
            background: #38bdf8;
            color: black;
            font-weight: bold;
            cursor: pointer;
        }

        .result {
            margin-top: 30px;
            background: #334155;
            padding: 20px;
            border-radius: 10px;
        }
    </style>
</head>
<body>

<div class="container">
    <h1>NeuroSync AI</h1>
    <p>Predictive Healthcare & Behavioral Drift Detection</p>

    <form method="POST">

        <label>Sleep Hours</label>
        <input type="number" step="0.1" name="sleep" required>

        <label>Heart Rate</label>
        <input type="number" name="heart_rate" required>

        <label>Stress Level</label>
        <select name="stress">
            <option>Low</option>
            <option>Medium</option>
            <option>High</option>
        </select>

        <label>Daily Steps</label>
        <input type="number" name="steps" required>

        <label>Screen Time (hours)</label>
        <input type="number" step="0.1" name="screen_time" required>

        <button type="submit">Analyze Behavior</button>

    </form>

    {% if result %}
    <div class="result">
        <h2>{{ result.prediction }}</h2>

        <h3>Detected Behavioral Drift:</h3>

        <ul>
            {% for item in result.anomalies %}
            <li>{{ item }}</li>
            {% endfor %}
        </ul>
    </div>
    {% endif %}
</div>

</body>
</html>
"""

@app.route("/", methods=["GET", "POST"])
def index():
    result = None

    if request.method == "POST":
        sleep = float(request.form["sleep"])
        heart_rate = int(request.form["heart_rate"])
        stress = request.form["stress"]
        steps = int(request.form["steps"])
        screen_time = float(request.form["screen_time"])

        risk_score = 0
        anomalies = []

        if sleep < 5:
            risk_score += 2
            anomalies.append("Low sleep duration detected")

        if heart_rate > 100:
            risk_score += 2
            anomalies.append("Elevated heart rate pattern")

        if stress == "High":
            risk_score += 3
            anomalies.append("High stress behavior detected")

        if steps < 3000:
            risk_score += 1
            anomalies.append("Low physical activity")

        if screen_time > 8:
            risk_score += 1
            anomalies.append("Excessive screen exposure")

        if risk_score >= 6:
            prediction = "High Health Risk"
        elif risk_score >= 3:
            prediction = "Moderate Health Risk"
        else:
            prediction = "Low Health Risk"

        result = {
            "prediction": prediction,
            "anomalies": anomalies
        }

    return render_template_string(HTML, result=result)

def run():
    app.run()

threading.Thread(target=run).start()

public_url = ngrok.connect(5000)
print("Your NeuroSync AI App Link:")
print(public_url)
```
